```python
import pandas as pd
taxi = pd.read_csv('https://stepik.org/media/attachments/lesson/359240/taxi_peru.csv', parse_dates=['start_at', 'end_at','arrived_at'],sep=';')
```

```python
taxi
```

```python
platform = taxi.groupby('source',as_index=False).count() \
    .sort_values('journey_id',ascending=False) \
    .head(1)
final = (platform['journey_id'] / taxi.shape[0]) * 100
round(final)
```

```python
scores = taxi \
    .groupby('driver_score',as_index=False) \
    .count()
scores = scores[['driver_score','journey_id']]
scores['journey_id'] = scores['journey_id'] / scores['journey_id'].sum()
scores['journey_id'] = scores['journey_id'].mul(100).round(2)
scores = scores.reset_index(drop=True)
scores = scores.rename(columns={'journey_id':'percentage'})
driver_score_counts = scores.sort_values('driver_score',ascending=True)
```

```python
driver_score_counts
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ax = sns.barplot(x='driver_score', y='percentage', data=driver_score_counts, color='blue', alpha=0.5)
ax.set(xlabel='Driver score', ylabel='Percentage') # - указываем названия столбцов
sns.despine()  # убрать часть рамки графика
print('Оценки клиентов')
plt.show()
```

```python
rider_score = taxi.groupby('rider_score',as_index=False).count()
rider_score = driver_score[['rider_score','journey_id']]
rider_score['journey_id'] = (driver_score['journey_id'] / driver_score['journey_id'].sum()) * 100
rider_score['journey_id'] = rider_score['journey_id'].round(2)
rider_score = rider_score.sort_values('rider_score',ascending=True)
rider_score = rider_score.rename(columns = {'journey_id':'percentage'})

ax = sns.barplot(x = 'rider_score', y = 'percentage', data = rider_score, color = 'blue', alpha = 0.5)
ax.set(xlabel='Rider score',ylabel='Percentage')
sns.despine()
print('Оценки водителей')
plt.show
```
